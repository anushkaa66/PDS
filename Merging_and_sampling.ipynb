```python
import pandas as pd

url1 = "/content/PDS/winequality-red.csv"
url2 = "/content/PDS/winequality-white.csv"

dataset1 = pd.read_csv(url1)
dataset2 = pd.read_csv(url2)

display(dataset1.head())
display(dataset2.head())

concat_dataset= pd.concat([dataset1, dataset2])
display(concat_dataset.head())

sample_dataset1 = dataset1.sample(n=5, random_state=1)
sample_dataset2 = dataset2.sample(n=5, random_state=1)

display(sample_dataset1)
display(sample_dataset2)

concat_sample = pd.concat([sample_dataset1, sample_dataset2])
display(concat_sample)
```
